```python
import nltk
import string
nltk.download('punkt')
```

```python
text = """
Natural Language Processing (NLP) is a part of Artificial Intelligence! It helps computers
understand human language. NLP is used in chatbots, translation systems, and voice assistants.

Students are learning Python, Java, and other languages. They practice coding daily; sometimes
they make mistakes, but learning never stops! Is programming interesting?
"""
```

```python
print("Original Text:\n")
print(text)
```

```python
text_remove_punctuation = text.translate(str.maketrans('', '', string.punctuation))

print(text_remove_punctuation)
```

```python
nltk.download('stopwords')
nltk.download('punkt_tab')
nltk.download('wordnet')
from nltk.corpus import stopwords

words = nltk.word_tokenize(text_remove_punctuation)

stop_words = set(stopwords.words("english"))
remaining_words = []

for word in words:
    if word.lower() not in stop_words:
        remaining_words.append(word)

print(remaining_words)
```

```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
stemmed_words = []
for word in remaining_words:
    stemmed_words.append(stemmer.stem(word))

print(stemmed_words)
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

lemmatized_words = []
for word in remaining_words:
    lemmatized_words.append(lemmatizer.lemmatize(word))


print(lemmatized_words)
```
